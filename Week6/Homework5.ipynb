## Problem 1

```rust
fn main(x: field, y: field) -> field {
  return 5*x**3 - 4*y**2*x**2 + 13*x*y**2 + x**2 - 10*y
}
```

$$
\begin{align*}
v1 &= x * x    \\
v2 &= v1 * x    \\
v3 &= y * y    \\
v4 &= v1 * v3    \\
v5 &= x * v3    \\
out &= (5*v2 - 4*v4 + 13 * v5 + v1 - 10 * y) * 1    \\

\end{align*}
$$

Witness: [1, x, y, out, v1, v2, v3, v4, v5]

```rust
fn main(x: field, y: field) -> field {
  assert!(y == 0 || y == 1 || y == 2); // y(y-1)(y-2) = 0    
    if (y == 0) { // 1/2 * y^2 - 3/2 * y + 1 
	    return x;
	} else if (y == 1) { // 2*y -y^2
	    return x**2;
	} else { // 1/2*y^2 - 1/2*y
	    return x**3;
	}
}
```

\end{align*}
